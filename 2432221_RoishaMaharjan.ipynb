```python
#task1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Copy of WHR-2024-5CS037.csv")  # Updated with correct file path

# Display the first 10 rows
print("First 10 rows of the dataset:")
print(df.head(10))

# Identify rows, columns, and data types
print("\nDataset Overview:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\nData Types:")
print(df.dtypes)

# Calculate mean, median, and standard deviation for the score column
print("\nStatistics for 'score' column:")
mean_score = df['score'].mean()
median_score = df['score'].median()
std_score = df['score'].std()
print(f"Mean: {mean_score}, Median: {median_score}, Standard Deviation: {std_score}")

# Identify the countries with the highest and lowest happiness scores
max_score_country = df[df['score'] == df['score'].max()]['Country name'].values[0]
min_score_country = df[df['score'] == df['score'].min()]['Country name'].values[0]
print(f"\nCountry with highest happiness score: {max_score_country}")
print(f"Country with lowest happiness score: {min_score_country}")

# Check for missing values and display counts if any
print("\nMissing Values:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

# Filter countries with scores > 7.5 and sort by Log GDP per capita in descending order
filtered_sorted_df = df[df['score'] > 7.5].sort_values(by='Log GDP per capita', ascending=False)
print("\nCountries with scores > 7.5 sorted by Log GDP per capita:")
print(filtered_sorted_df[['Country name', 'score', 'Log GDP per capita']])

# Add a "Happiness Category" based on score thresholds
def happiness_category(score):
    if score >= 7.5:
        return 'Very Happy'
    elif score >= 5.0:
        return 'Moderately Happy'
    else:
        return 'Unhappy'

df['Happiness Category'] = df['score'].apply(happiness_category)
print("\nHappiness Categories added:")
print(df[['Country name', 'score', 'Happiness Category']].head())

# Data Visualization
# 1. Bar plot of top 10 happiest countries
plt.figure(figsize=(10, 6))
top10 = df.nlargest(10, 'score')
sns.barplot(x='score', y='Country name', data=top10, palette='viridis')
plt.title('Top 10 Happiest Countries')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()

# 2. Line plot of scores by Log GDP per capita
plt.figure(figsize=(10, 6))
sns.lineplot(x='Log GDP per capita', y='score', data=df)
plt.title('Happiness Score vs Log GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# 3. Histogram of Happiness Scores
plt.figure(figsize=(10, 6))
plt.hist(df['score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()

# 4. Scatter plot of Log GDP per capita vs score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', hue='Happiness Category', data=df, palette='coolwarm')
plt.title('Happiness Score vs Log GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend(title='Happiness Category')
plt.show()



```
